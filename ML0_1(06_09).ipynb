Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.


```python
def func(N, f_el, d):
  diag_ = np.arange(f_el, f_el+d*(N-1)+1, d)
  return np.diag(diag_)
func(5, 1, 3)
```

Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
matrix1 = np.array([5 if i%2 == 0 else 0 for i in range(25)]).reshape(5,5)
matrix2 =  np.array([[0 if i<j else 5 for i in range(5)] for j in range(5)])

print(matrix1, '\n')
print(matrix2, '\n')

print(f'Детерминанты matrix1 = {np.linalg.det(matrix1)} matrix2 = {np.linalg.det(matrix2)}')
print(f'Обратная матрица\n{np.linalg.inv(matrix2)}')
```

17. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data_ = data.groupby("Rooms")['Rooms'].count()
explode = [0 if i != max(data_.values) else 0.1 for i in data_.values]
legend = ['{} ({}%)'.format(int(idx), (value*100)/10000) for idx, value in zip(data_.index, data_.values)]
```

```python
ax = data_.plot.pie(explode = explode, labeldistance=None)
ax.legend(loc='best', labels = legend)
plt.show()
```
