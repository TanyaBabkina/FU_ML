# Задание 11

11. Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В
, ... Вк
 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.

```python
for i in range(shortAns):
  df.insert(len(df.columns), f'B{i+1}', 0)

for i, _ in df.iterrows():
  shortAns = df.loc[i, 'Задания с кратким ответом']
  for j in range(len(shortAns)):
    df.loc[i, f'B{j+1}'] = shortAns[j]

cols = df.loc[:, 'B1':'B12'].columns
for col in cols:
  print(col, df.groupby([col]).agg({col:'count'})/df.shape[0]*100)
```

# Задание 12

12. Аналогично и с типом С (ответы с развернутым ответом)

```python
for i in range(longAns):
  df.insert(len(df.columns), f'C{i+1}', 0)

for i, _ in df.iterrows():
  c = 0
  for j in range(0, len(df['Задания с развёрнутым ответом'][1]),4):
    longAns = df.loc[i, "Задания с развёрнутым ответом"]
    df.loc[i, f'C{c+1}'] = longAns[j]
    c+=1
```

```python
schools = df[(df['№ школы'] == 132)|(df['№ школы'] == 136)]

tasks = list(schools.loc[:,'B1':'B12'])

percents_136 = [len(df[(df['№ школы'] == 136) & (df[i] == '+')])/len(df[df['№ школы'] == 136])*100 for i in tasks]
percents_132 = [len(df[(df['№ школы'] == 132) & (df[i] == '+')])/len(df[df['№ школы'] == 132])*100 for i in tasks]

new_df= pd.DataFrame(index = tasks, data = {'School_132': percents_132, 'School_136':percents_136})
new_df.plot.bar()
```

```python
import re
max_score = [round(int(i)/2) for i in re.findall(r'\((\d)\)', df['Задания с развёрнутым ответом'][0])]
tasks = list(schools.loc[:,'C1':'C7'])
task_score = zip(tasks, max_score)

percents_136 = [len(df[(df['№ школы'] == 136) & (df[task] >= str(max))])/len(df[df['№ школы'] == 136])*100 for task, max in zip(tasks, max_score)]
percents_132 = [len(df[(df['№ школы'] == 132) & (df[task] >= str(max))])/len(df[df['№ школы'] == 132])*100 for task, max in zip(tasks, max_score)]

new_df= pd.DataFrame(index = tasks, data = {'School_132': percents_132, 'School_136':percents_136})
new_df.plot.bar()
```
